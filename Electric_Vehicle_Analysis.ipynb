# Predictive Analysis of the European Electric Vehicle Market: Infrastructure & Performance Trends

## About the Dataset

**Context:**
This dataset offers a comprehensive overview of the Electric Vehicle (EV) market in Germany, incorporating technical specifications, performance metrics, and market pricing based on 2024 data for a diverse set of EV models from leading global manufacturers

**Purpose:**
The dataset is intended to analyze the relationship between battery technology, vehicle performance (range and acceleration), and market value. It facilitates exploratory data analysis to uncover patterns, correlations, and insights into EV pricing dynamics and factors influencing consumer adoption within the European energy transition framework.

**Dataset Features:**
The dataset contains several key parameters essential for automotive and energy sector analysis:

**Battery**: Useable battery capacity (in kWh).

**Efficiency**: Energy consumption (in Wh/km).

**Fast Charge**: Fast charging speed (in km/h).

**Price.DE**: Current market price in Germany (in Euro). Includes "NA" values for data cleaning practice.

**Range**: Maximum travel distance on a single charge (in km).

**Top Speed**: Maximum vehicle speed (in km/h).

**Acceleration** (0-100): Time taken to reach 100 km/h (in seconds).

**Car Name & Link**: Categorical strings identifying the model and source data.

### Dataset Scope

The dataset is specifically curated from a German market perspective, making it highly relevant for applications to German universities or automotive firms.

The data reflects the diverse landscape of the EV industry from budget-friendly city cars (like the Dacia Spring) to ultra-luxury performance models (like the Rolls-Royce Spectre). This variety allows for a nuanced exploration of how different segments of the market prioritize efficiency versus performance.

### Feature Engineering Opportunity:

**Price per Range**: A derived metric to calculate the value-for-money ratio of different models.

**Charging Efficiency**: A derived metric combining Battery size and Fast Charge speed to determine long-distance travel viability.

## Data Cleaning and Preprocessing

```python
# Dropping the link column as it is not required for statistical analysis

ev_df.drop(columns=['Car_name_link'], inplace=True)
```

```python
ev_df.tail()
```

```python
# Renaming specific columns

ev_df.rename(columns={
    'Price.DE.': 'Price_DE',
    'acceleration..0.100.': 'Acceleration_0_100'
}, inplace=True)
```

```python
#to display last 5 rows of dataframe

ev_df.tail()
```

```python
# Returns the count of NaN values for each column

ev_df.isnull().sum()
```

```python
# Filling Fast_charge with the overall median

ev_df['Fast_charge'] = ev_df['Fast_charge'].fillna(ev_df['Fast_charge'].median())
```

```python
ev_df.isnull().sum()
```

```python
# Create the 'Manufacturer' column by taking the first word of 'Car_name'
ev_df['Manufacturer'] = ev_df['Car_name'].str.split().str[0]

# Check the first few rows to confirm it worked
ev_df[['Car_name', 'Manufacturer']].head()
```

```python
# Filling Price_DE using two-tier imputation strategy

# Tier 1: Brand-Specific Imputation
# Fill NaNs using the median price of other cars from the same Manufacturer
brand_medians = ev_df.groupby('Manufacturer')['Price_DE'].transform('median')
ev_df['Price_DE'] = ev_df['Price_DE'].fillna(brand_medians)

# Tier 2: Global Median Fallback (Full Coverage)
# If a brand only has one car and its price is missing, Tier 1 won't work.
# This line catches any remaining NaNs using the overall dataset median.
ev_df['Price_DE'] = ev_df['Price_DE'].fillna(ev_df['Price_DE'].median())
```

```python
ev_df.isnull().sum()
```

```python
ev_df.info()
```

```python
# Standardizing all numeric columns to float64

ev_df['Efficiency'] = ev_df['Efficiency'].astype(float)
ev_df['Range'] = ev_df['Range'].astype(float)
ev_df['Top_speed'] = ev_df['Top_speed'].astype(float)
```

Individual circles on the right side (beyond €110,000) means the dataset includes both "mass-market" cars and "luxury/hyper" EVs.

```python
# Create a 'Price_per_km' column
#Since analyzing the European market, creating a "Value for Money" metric. This shows how many Euros you pay for every 1 km of range.

ev_df['Price_per_km'] = ev_df['Price_DE'] / ev_df['Range']
```

The EV market in Germany is primarily dominated by mid-range vehicles, with most models priced between 40,000 and 70,000. The median price reinforces this central concentration. While high-end luxury EVs exist, they form a smaller portion of the market, resulting in a right-skewed distribution. This suggests that EV adoption is largely centered around moderately priced vehicles, making them more accessible compared to premium alternatives.

**Answer:**
The German EV market is primarily dominated by mid-range vehicles, indicating a transition from a niche luxury market to a more accessible, broader consumer segment. While a "long tail" of high-performance luxury models exists, the core market volume is concentrated in the €40,000 to €70,000 price bracket.

#### Identifying Luxury Segment Outliers in Price vs. Range

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = ev_df['Price_DE'].quantile(0.25)
Q3 = ev_df['Price_DE'].quantile(0.75)

# Calculate the Interquartile Range
IQR = Q3 - Q1

# Define the Upper Fence (1.5 is the standard multiplier)
upper_fence = Q3 + 1.5 * IQR

print(f"The 'Normal' Price Limit is: €{upper_fence:,.2f}")
```

```python
# Filter the dataframe to see only the outliers
luxury_outliers = ev_df[ev_df['Price_DE'] > upper_fence]

# Display the top outliers sorted by price
print(luxury_outliers[['Manufacturer', 'Car_name', 'Price_DE']].sort_values(by='Price_DE', ascending=False))
```

```python
# Create a filtered dataframe that excludes these luxury outliers
ev_df_mass_market = ev_df[ev_df['Price_DE'] <= upper_fence]

# Plot the "Clean" Bivariate Analysis
plt.figure(figsize=(10, 6))
sns.regplot(data=ev_df_mass_market, x='Range', y='Price_DE', 
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

plt.title('Price vs. Range: Mass Market Trend (Excluding Luxury Outliers)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

#### Comparison: Full Market vs. Mass Market Dynamics
Key Observations:

- Standardization: By placing Range (Independent Variable) on the X-axis, the model more logically predicts Price (Dependent Variable).
- Variance Reduction: Filtering the luxury outliers (Price > €115k) significantly reduces the variance.

Conclusion: The refined plot shows a much stronger linear relationship, indicating that for the majority of German EV consumers, range remains the primary determinant of vehicle cost.

**Answer:**
The refined analysis reveals a strong, consistent linear relationship between price and driving range in the mass-market segment. Once luxury outliers are removed, the data shows that range is the primary cost driver for the majority of EVs in Germany. The previous 'significant variation' is greatly reduced, proving that for non-luxury models, price and range are strictly correlated. Consumers effectively pay a standardized premium for every additional kilometer of battery capacity.

#### 4.1 Efficiency vs. Manufacturer

**Research Question:** To what extent does vehicle efficiency ($Wh/km$) act as a mediating factor between battery capacity and real-world range across different manufacturer segments?

```python
# Calculate medians and get the top/bottom 10
medians = ev_df.groupby('Manufacturer')['Efficiency'].median().sort_values()
top_bottom_brands = pd.concat([medians.head(10), medians.tail(10)]).index

# Filter the dataframe
# Keeps only data for those 20 selected manufacturers
filtered_df = ev_df[ev_df['Manufacturer'].isin(top_bottom_brands)]

# Plot the filtered version
plt.figure(figsize=(10, 6)) 
sns.boxplot(data=filtered_df, x='Efficiency', y='Manufacturer', order=top_bottom_brands, hue='Manufacturer', legend=False, palette='RdYlGn_r')
plt.title('The Efficiency Gap: Top 10 vs Bottom 10 Manufacturers')
```

#### Understanding the Plot

This horizontal box plot visualizes the distribution of energy efficiency across different electric vehicle brands, highlighting the performance disparity between market leaders and luxury or performance-heavy manufacturers.

- The x-axis represents the Efficiency (Wh/km), where a lower value indicates a more energy-efficient vehicle.
- The y-axis represents the Manufacturer names, categorized into the top and bottom performers in the dataset.
- Central Tendency: The vertical line inside each box represents the median efficiency for that specific brand.
- Efficiency Leaders: Brands like Dacia and Tesla are positioned at the top of the chart with the lowest Wh/km values (approx. 140–160 Wh/km), showing   superior energy conservation.
- High-Variance Brands: Manufacturers like Fiat, Peugeot, and Opel show very wide boxes, indicating a diverse fleet that ranges from highly efficient    city cars to much less efficient models.
- The Luxury "Tax": Brands like Lotus, Maserati, and Hongqi cluster at the bottom with significantly higher consumption (above 220 Wh/km), reflecting    the energy demands of high-performance motors and heavier luxury chassis.
- Outliers: The small circles (notably in the Tesla and Renault rows) represent specific models that fall significantly outside the brand's typical      efficiency range.
    
**Inference:**  
The data reveals a clear efficiency divide: mass-market and city-car brands achieve the best economy, while luxury and performance manufacturers sacrifice efficiency for power. This results in a significant 40–50% efficiency gap across the market, highlighting inconsistent optimization levels between different brand portfolios.
    
**Answer:**
Vehicle efficiency acts as a regulator of the battery-to-range relationship. In the Efficiency-First segment (Dacia, Renault), it compensates for small batteries to provide adequate range. In the Performance segment, poor efficiency acts as a "drain," requiring much larger batteries to achieve the same range as more efficient, mid-tier models. Therefore, while battery capacity sets the potential range, efficiency defines the actualized range achieved per unit of energy.

#### 8. Price per km vs Manufacturer

**Research Question:** How does the average Price_per_km compare across the Top 10 manufacturers in the 2024 German market, and which brand offers the most "range for the Euro"?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Filter for Top 10 Manufacturers by vehicle count
top_10_manufacturers = ev_df['Manufacturer'].value_counts().nlargest(10).index
df_top10 = ev_df[ev_df['Manufacturer'].isin(top_10_manufacturers)]

# 2. Calculate the mean Price_per_km for these Top 10
brand_value = df_top10.groupby('Manufacturer')['Price_per_km'].mean().sort_values()

# 3. Visualization
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Using a sequential palette to highlight the "Value Leaders" (lowest Price_per_km)
ax = sns.barplot(
    x=brand_value.values, 
    y=brand_value.index, 
    hue=brand_value.index,  # Assign y to hue
    palette="viridis_r", 
    legend=False            # Hide the legend as it's redundant here
)

# Customizing for clarity and aesthetic contrast
plt.title('Average Price per km by Top 10 Manufacturers (2024 German Market)', pad=20)
plt.xlabel('Average Price per km (€/km)', fontsize=12)
plt.ylabel('Manufacturer', fontsize=12)

# Adding data labels for precision
for i, v in enumerate(brand_value.values):
    ax.text(v + 0.5, i, f'€{v:.2f}', color='black', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

#### Correlation Matrix : Interdependence of EV Specifications

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the Pearson correlation matrix
# (Since all your columns are floats, this will be very accurate)
corr_matrix = ev_df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(10, 7))

# Create the heatmap
# 'annot=True' displays the numbers in the cells
# 'cmap='coolwarm'' uses Blue for negative and Red for positive correlations
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap: Interdependence of EV Specifications')
plt.show()
```

#### Understanding the Correlation Heatmap

This analysis breaks down the Correlation Heatmap to understand how various EV specifications—like battery, range, and price—interact with one another.

- The x-axis and y-axis represent the same set of EV specifications, creating a symmetric matrix.
- The color scale (coolwarm) indicates the strength of the relationship: 1.0 (dark red) is a perfect positive correlation, 0.0 (white) is no correlation, and -1.0 (dark blue) is a perfect inverse correlation.
- Strong positive clusters are visible between Battery, Range, and Top Speed, indicating that high-performance EVs typically require larger battery capacities.
- Negative correlations are most prominent with Acceleration 0-100. Because a "lower" time is better, the blue cells show that higher top speeds and larger batteries lead to faster (lower numerical) acceleration times.
- Efficiency shows the lowest overall correlation with other metrics (near 0.0), suggesting it is an independent design factor not strictly tied to price or power.
- Price_DE (Price in Germany) shows a strong tie to Top Speed (0.71) and Battery (0.67), confirming these are the primary cost drivers in the current market.
- Range and Battery (0.88) represent the strongest linear bond in the dataset, proving that battery size remains the most critical factor for distance, overshadowing motor efficiency.
  
    
**Inference:**  
The heatmap shows that the EV market is mainly driven by three factors: battery size, performance, and price. Cars with larger batteries usually have longer range and better performance (speed and acceleration), which makes them more expensive. Interestingly, efficiency is not strongly related to price, suggesting that both manufacturers and consumers focus more on performance and range rather than energy efficiency. This means that battery size and top speed are the most useful factors for estimating an EV’s price.
